```python
# today's date.
today = date.today()
# Take Date of Birth from user
dob = input("Enter your Date of Birth (DD/MM/YYYY): ")

d, m, y = map(int, dob.split("/"))

# Calculate age of the user
years = today.year - y
months = today.month - m
days = today.day - d

# Simple adjustments
if days < 0:
    months = months - 1
    days = days + 30

if months < 0:
    years = years - 1
    months = months + 12

# Display result
print("Your Age is:")
print(years, "Years")
print(months, "Months")
print(days, "Days")
```
